```python
from transformers import pipeline
```

```python
a=pipeline("sentiment-analysis")
```

```python
gen=pipeline("text-generation")
```

```python
text=input("Enter the text:")
t=(gen(text,max_length=200))
```

```python
print(t[0].get('generated_text'))
```

```python
n=pipeline('ner')
```

```python
text=input("Enter the text:")
result = n(text)
result
```

```python
qa=pipeline("question-answering")
```

```python
c = '''Bitty bought some butter,but the butter was itter butter, So, Bitty bought some better butter, to make the butter better.'''
```

```python
q="What does Bitty bought?"
```

```python
result = qa({'question':q, 'context' : c})
print(result)
```
